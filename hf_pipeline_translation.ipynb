# Пример использования библиотеки Hugging Face для автоматического перевода

Устанавливает библиотеку Hugging Face (модуль transformers)

```python
!pip install transformers sentencepiece
```

Создаем пайплайн для создания автоматического перевода

```python
from transformers import pipeline

translator = pipeline("translation_ru_to_en", "Helsinki-NLP/opus-mt-ru-en")
```

Используем пайплайн для создания перевода


```python
translator("Да прибудет с тобой сила, студент")
```
